```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Load your dataset into a pandas DataFrame
data = pd.read_csv('newdata.csv')

# Separate the features and the target variable
X = data.drop('Target', axis=1)  # Replace 'target_variable' with your actual target column name
y = data['Target']

# Create a random forest classifier
rf = RandomForestClassifier()

# Fit the classifier to the data
rf.fit(X, y)

# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame with feature names and their importance scores
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the features by importance in descending order
feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

# Set a threshold for selecting important features (e.g., top 10%)
threshold = 0.45
num_features = int(threshold * len(feature_importance_df))

# Get the top features above the threshold
top_features = feature_importance_df.head(num_features)['Feature']

# Print the top features
print("Top features:")
for feature in top_features:
    print(feature)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from skrebate import ReliefF


# Separate the features and the target variable
X = student.drop('Target', axis=1)
y = student['Target']

# Perform Relief feature selection
relief = ReliefF()
relief.fit(X.values, y.values)

# Get the feature importance scores
feature_scores = relief.feature_importances_

# Sort the features based on their importance scores
top_features_indices = np.argsort(feature_scores)[::-1]

# Select the top-ranked features from the dataset
selected_features = X.columns[top_features_indices[:15]]  # Adjust the number of selected features as needed

# Create a new dataframe with the selected features
selected_data = X[selected_features]
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler



# Separate the features (X) and the target (y)
X = student_df.drop('Target', axis=1)
y = student_df['Target']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_var_ratio = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_var = np.cumsum(explained_var_ratio)

# Print the explained variance ratio and cumulative explained variance
for i, var_ratio in enumerate(explained_var_ratio):
    print(f"PC{i+1} Explained Variance Ratio: {var_ratio:.2f}")
    print(f"PC{i+1} Cumulative Explained Variance: {cumulative_explained_var[i]:.2f}")

# Plotting the explained variance ratio
plt.plot(range(1, len(explained_var_ratio) + 1), cumulative_explained_var, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Principal Components')
plt.show()
```

```python
X = selected_data.iloc[:,0:13]
y = selected_data.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, f1_score, mean_squared_error
clf = LogisticRegression()

# Without Scaling 
clf.fit(X_train,y_train) 
y_pred = clf.predict(X_test)
print("Without Scaling and without CV - Accuracy Score: ", accuracy_score(y_test, y_pred))
print("Without Scaling and without CV - Recall Score: ", recall_score(y_test, y_pred, average='macro'))
print("Without Scaling and without CV - F-score: ", f1_score(y_test, y_pred,average='macro'))
print("Without Scaling and without CV - MSE: ", mean_squared_error(y_test, y_pred))
scores = cross_val_score(clf, X_train, y_train, cv=10)
print("Without Scaling and With CV: ",scores.mean())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Assuming you have already trained your logistic regression model and obtained the predicted probabilities
y_pred_prob = clf.predict_proba(X_test)

# Compute the false positive rate, true positive rate, and threshold for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
num_classes = 3  # Number of classes in your multiclass problem
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test, y_pred_prob[:, i], pos_label=i)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot the ROC curves for each class
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green']  # Customize the colors based on the number of classes
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

# Unsupervised Learning :

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import davies_bouldin_score, silhouette_score
import matplotlib.pyplot as plt

# Load and preprocess the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(selected_data)

# Perform hierarchical clustering
linkage_matrix = linkage(data_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(40, 40))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Determine the number of clusters
silhouette_scores = []
max_clusters = len(data_scaled) - 1

for num_clusters in range(2, max_clusters + 1):
    clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
    silhouette_scores.append(silhouette_score(data_scaled, clusters))

best_num_clusters = np.argmax(silhouette_scores) + 2  # Add 2 to start counting from 2 clusters
print("Best number of clusters:", best_num_clusters)

# Assign the clusters
clusters = fcluster(linkage_matrix, best_num_clusters, criterion='maxclust')

# You can also analyze the characteristics of each cluster using the assigned cluster labels
selected_data['Cluster'] = clusters
cluster_stats = selected_data.groupby('Cluster').mean()
print(cluster_stats)
```
